```python
import pandas as pd

feature_names = ['party', 'handicapped-infants', 'water-project-cost-sharing',
                 'adoption-of-the-budget-resolution', 'physician-fee-freeze',
                 'el-salvador-aid', 'religiouss-groups-in-schools',
                 'anti-satellite-test-ban', 'aid-to-nicaraguan-contras',
                 'mx-missle', 'immigration', 'synfuels-corporation-cutback',
                 'education-spending', 'superfund-right-to-sue', 'crime',
                 'duty-free-exports', 'export-administration-act-south-africa'] 

voting_data = pd.read_csv('house-votes-84.data.txt', na_values=['?'],
                          names=feature_names)


voting_data.head()
```

```python
voting_data.describe()
```

```python
voting_data.dropna(inplace=True)
voting_data.describe()
```

```python
voting_data.replace(('y', 'n'), (1, 0), inplace=True)
voting_data.replace(('democrat', 'republican'), (1, 0), inplace=True)
```

```python
voting_data.head()
```

```python
all_features = voting_data[feature_names].values
all_classes = voting_data['party'].values
```

```python
voting_data
```

```python
all_features
```

```python
all_classes
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(all_features, all_classes, test_size=0.2, random_state=42)
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(8, activation='relu', input_shape=(17,)))
model.add(Dropout(0.5))
model.add(Dense(8, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train,
                   epochs=20,
                   verbose=2,
                   validation_data=(X_test, y_test))
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test acc:', score[1])
```
